## ex. 1

$$  
y = x^Tx,  \quad x \in \mathbb{R}^N
$$

$$
\frac{dy}{dx} = 2x
$$



```
Есть два подхода
1) расписать через сумму, посмотреть на производные слагаемых и вывести ответ в общем виде
2) вспомнить базовые формулы матричного дифференцирования

1) y = x^T x = ∑ᵢ xᵢ²
∂y/∂xⱼ = ∂/∂xⱼ (∑ᵢ xᵢ²) = 2xⱼ
dy/dx = 2x

2) По правилу для квадратичных форм: если y = x^T A x, то dy/dx = (A + A^T)x
В нашем случае A = I (y = x^T x можно записать как y = x^T I x), поэтому:
dy/dx = (I + I)x = 2Ix = 2x
```



## ex. 2

$$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $$

$$
\frac{dy}{dA} = B^T
$$

```
Распишем след произведения матриц:
y = tr(AB) = ∑ᵢ (AB)ᵢᵢ = ∑ᵢ ∑ₖ Aᵢₖ Bₖᵢ
Найдём частную производную по элементу Aⱼₘ:
∂y/∂Aⱼₘ = ∂/∂Aⱼₘ (∑ᵢ ∑ₖ Aᵢₖ Bₖᵢ)
Только слагаемое с i = j и k = m даёт ненулевой вклад:
∂y/∂Aⱼₘ = Bₘⱼ
Следовательно, матрица производных:
(dy/dA)ⱼₘ = Bₘⱼ
```

## ex. 4

Classic matrix factorization example. Given matrix $X$ you need to find $A$, $S$ to approximate $X$. This can be done by simple gradient descent iteratively alternating $A$ and $S$ updates.
$$
J = || X - AS ||_F^2  , \quad A\in \mathbb{R}^{N \times R} , \quad S\in \mathbb{R}^{R \times M}
$$
$$
\frac{dJ}{dS} = ?
$$

You may use one of the following approaches:

#### First approach
Using ex.2 and the fact:
$$
|| X ||_F^2 = tr(XX^T)
$$
it is easy to derive gradients (you can find it in one of the refs).

#### Second approach
You can use *slightly different techniques* if they suits you. Take a look at this derivation:
<img src="grad.png">
(excerpt from [Handbook of blind source separation, Jutten, page 517](https://books.google.ru/books?id=PTbj03bYH6kC&printsec=frontcover&dq=Handbook+of+Blind+Source+Separation&hl=en&sa=X&ved=0ahUKEwi-q_apiJDLAhULvXIKHVXJDWcQ6AEIHDAA#v=onepage&q=Handbook%20of%20Blind%20Source%20Separation&f=false), open for better picture).

#### Third approach
And finally we can use chain rule!
let $ F = AS $

**Find**
$$
\frac{dJ}{dF} = 2(F - X) = 2(AS - X)$$
and
$$
\frac{dF}{dS} = A_{ip} \delta_{jq}$$

(the shape should be $ NM \times RM )$.


$$
\frac{\partial J}{\partial S_{pq}} = \sum_{i,j} \frac{\partial J}{\partial F_{ij}} \frac{\partial F_{ij}}{\partial S_{pq}} = \sum_{i} 2(F_{iq} - X_{iq}) A_{ip} = 2\sum_{i} A_{ip}(F_{iq} - X_{iq})
$$

Now it is easy do get desired gradients:
$$
\frac{dJ}{dS} = 2A^T(F - X) = 2A^T(AS - X)
$$

### Question 2

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

*Your Answer:*
1, 2, 3

*Your Explanation:*
1. Subtracting the mean $\mu$ does not change the L1 distances between any pairs of points, as the same constant is subtracted from every feature in every sample 
2. Subtracting the per-pixel mean $\mu_{ij}$ also preserves L1 distances, since for each feature, the same value is subtracted across all samples
3. Subtracting $\mu$ and dividing by $\sigma$ scales all L1 distances by the same factor (1/$\sigma$), but since the scaling is uniform, the ordering of distances is preserved

1. The decision boundary (hyperplane between classes in feature space) of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

*Your Answer:*
2, 4

*Your Explanation:*

(2) Для 1-NN каждая обучающая точка будет классифицирована по самой себе (ближайший сосед - она сама), поэтому обучающая ошибка 1-NN = 0 (при отсутствии одинаковых точек с разными метками).
Для 5-NN обучающая точка классифицируется по 5 ближайшим соседям, что может привести к ошибкам, если среди соседей есть точки других классов.

(4) k-NN не строит явную модель на этапе обучения.
При классификации нового примера алгоритм должен:
- Вычислить расстояние до всех точек обучающей выборки
- Найти k ближайших соседей
- Выполнить голосование

Временная сложность классификации одного примера: O(N·d), где N - размер обучающей выборки, d - размерность признаков.
Следовательно, время растет линейно с размером обучающей выборки.
